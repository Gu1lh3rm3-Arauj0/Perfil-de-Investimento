#Algoritmo KNN

```python
def distancia_euclidiana(lista1, lista2):
    soma = 0

    #Distância de manhattan
    for i in range(len(lista1)):
      if (lista1[i] - lista2[i]) < 0:
          soma += -(lista1[i] - lista2[i])
      else:
          soma += (lista1[i] - lista2[i])

    return soma

def prever(X_treino, y_treino, x_teste, k):
    distancias = [distancia_euclidiana(x_teste, x) for x in X_treino]

    k_indices = sorted(range(len(distancias)), key=lambda i: distancias[i])[:k]

    k_rotulos_mais_comuns = []
    for i in k_indices:
      k_rotulos_mais_comuns.append(*y_treino[i])

    resultado = []
    resultado.append(statistics.mode(k_rotulos_mais_comuns))
    return resultado

def vizinhos(k, X_treino, y_treino, X_teste):
    previsoes = [prever(X_treino, y_treino, x, k) for x in X_teste]
    return previsoes
```

#Funções

```python
def separa_X(*data):
  X = []

  for dados in data:
    lista = []
    lista.append(dados[2][0])
    lista.append(dados[2][1])
    lista.append(dados[2][2])
    lista.append(dados[2][3])
    X.append(lista)

  return X
```

```python
def separa_Y(*data):
  Y = []

  for dados in data:
    lista = []
    lista.append(dados[1])

    Y.append(lista)

  return Y
```
